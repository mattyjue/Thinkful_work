```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap
from sklearn.pipeline import Pipeline
import seaborn as sns
```

# 1. Load the dataset and conduct any necessary preprocessing, such as normalizing the data.

```python
mnist = fetch_openml('Fashion-MNIST', version=1, cache=True)
```

```python
np.random.seed(123)

indices = np.random.choice(70000,10000)
X = mnist.data[indices] / 255.0 # Divide by 255 to normalize for RGB (Red, Green, Blue) in image data
y = mnist.target[indices]

print(X.shape, y.shape)
```

# 2. Apply UMAP to the data.

```python
umap_results = umap.UMAP(n_neighbors=5,
                      min_dist=0.3,
                      metric='correlation').fit_transform(X)
```

# 3. Using the two-dimensional UMAP representation, draw a graph of the data by coloring and labeling the data points as you did in the checkpoint.

```python
umap_components=pd.DataFrame(umap_results, columns=["umap1", "umap2"])
sns.scatterplot(x='umap1', y='umap2', data=umap_components)
plt.show()

color_plot=umap_components.copy()
color_plot['type']=y
sns.scatterplot(x='umap1', y='umap2', hue='type', alpha=.2,data=color_plot)
plt.show()
```
