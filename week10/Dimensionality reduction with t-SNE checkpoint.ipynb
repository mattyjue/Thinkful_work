```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import time
import seaborn as sns
```

# 1. Load the dataset and conduct any necessary preprocessing, such as normalizing the data.

```python
mnist = fetch_openml('Fashion-MNIST', version=1, cache=True)
```

```python
np.random.seed(123)

indices = np.random.choice(70000,10000)
X = mnist.data[indices] / 255.0 # Divide by 255 to normalize for RGB (Red, Green, Blue) in image data
y = mnist.target[indices]

print(X.shape, y.shape)
```

# 2. Apply t-SNE to the data by setting n_components=2.

```python
time_start = time.time()
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(X)

print('t-SNE done! Time elapsed: {} seconds'.format(time.time()-time_start))
```

# 3. Using the two-dimensional t-SNE representation, draw a graph of the data by coloring and labeling the data points as you did in the checkpoint.

```python
tse_components=pd.DataFrame(tsne_results, columns=["tsne1", "tsne2"])
sns.scatterplot(x='tsne1', y='tsne2', data=tse_components)
plt.show()

color_plot=tse_components.copy()
color_plot['type']=y
sns.scatterplot(x='tsne1', y='tsne2', hue='type', alpha=.2,data=color_plot)
plt.show()
```

# 5. Now, play with the different perplexity values and apply t-SNE for each of them. Which perplexity value is best in terms of the two-dimensional representation clarity?

```python
perps=[10, 20, 30, 50]

for p in perps:
    tsne = TSNE(n_components=2, verbose=1, perplexity=p, n_iter=300)
    tsne_results = tsne.fit_transform(X)
    tse_components=pd.DataFrame(tsne_results, columns=["tsne1", "tsne2"])
    sns.scatterplot(x='tsne1', y='tsne2', data=tse_components)
    plt.title(f'No color scatter with perplexity of {p}')
    plt.show()

    color_plot=tse_components.copy()
    color_plot['type']=y
    sns.scatterplot(x='tsne1', y='tsne2', hue='type', alpha=.2,data=color_plot)
    plt.title(f'Color scatter with perplexity of {p}')
    plt.show()
```

All the groupings had similar shapes but rotated in different ways. All the perplexities were similar in two dimensional clarity.
