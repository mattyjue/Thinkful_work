```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import seaborn as sns
```

# 1. Load the dataset and conduct any necessary preprocessing, such as normalizing the data.

```python
mnist = fetch_openml('Fashion-MNIST', version=1, cache=True)
```

```python
np.random.seed(123)

indices = np.random.choice(70000,10000)
X = mnist.data[indices] / 255.0 # Divide by 255 to normalize for RGB (Red, Green, Blue) in image data
y = mnist.target[indices]

print(X.shape, y.shape)
```

# 2. Apply PCA to the data and get the first two principal components.

```python
pca=PCA(2)
pca_components=pca.fit_transform(X)
pca_components = pd.DataFrame(pca_components, columns=["pca1", "pca2"])
pca_components.head()
```

```python
pca.explained_variance_ratio_
```

# 3. Using the first two principal components, draw a graph of the data by coloring and labeling the data points as you did in the checkpoint.

```python
sns.scatterplot(x='pca1', y='pca2', data=pca_components)
```

```python
color_plot=pca_components.copy()
color_plot['type']=y
sns.scatterplot(x='pca1', y='pca2', hue='type', alpha=.2,data=color_plot)
plt.show()
```
